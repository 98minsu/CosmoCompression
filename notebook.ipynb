# Linear Compression Methods

The first 5 methods presented here are linear comprsesion methods. The original data vector ($t$) is transformed to a compressed data vector ($c$) by multiplying it with a transformation matrix $U, \ c = Ut$.

How $U$ is constructed varies by method.

# PCA 

First is PCA. We take the simulated data vectors from varying cosmological parameters, and compute the covariance matrix and corresponding eigenvectors.

The 16 components with the highest variance are taken as the rows of the linear dimensionality reduction matrix.

There is a clear elbow to the eigenvalue plot around the 20th eigenvalue. The components up to the elbow can be understood as carying cosmological information, and the rest can be seen as noise.

The correlation matrix plots show that the components of the compressed eigenvalue are uncorrelated and there is good correlation with the parameters of interest

```python
# Compute the covariance matrix
cov = np.cov(dv.T)

# np's eigh returns lambda and vectors in increasing order of lambda
evals, evecs = np.linalg.eigh(cov)
evals = evals[::-1]
evecs = evecs[:,::-1]

plt.title("Eigenvalue Plot for PCA")
plt.xlabel("Eigenvalue Ranking")
plt.ylabel(r"$\lambda$")
plt.plot(evals)
plt.yscale('log')
plt.show()

# compress to 16 components like other methods
evecs = evecs[:,:16]

# compression is a linear operation
compressed_dv   =  np.array((dv_LFI)@evecs)

PCC = np.abs(np.corrcoef( compressed_dv,pars_LFI, rowvar = False))
makeplot(PCC, 'PCA')
```

# PCA-f

Second is PCA-f. We take the simulated data vectors from fixed fiducial cosmological parameters and compute the eigenvectors of the covariance matrix. 

Once again we only use the 16 components with the highest variance.

There is no elbow to the eigenvalue curve here. This is because the covariance matrix here only contains noise information, so there is no distinction between cosmological signal modes and noise modes.

The correlation matrix plots show that the components of the compressed eigenvalue are highly correlated and there is no significant correlation with the parameters of interest. 

```python
cov = np.cov(dv_cov.T)

# np's eigh returns lambda and vectors in increasing order of lambda
evals, evecs = np.linalg.eigh(cov)
evals = evals[::-1]
evecs = evecs[:,::-1]


plt.title("Eigenvalue Plot for PCA-f")
plt.xlabel("Eigenvalue Ranking")
plt.ylabel(r"$\lambda$")
plt.plot(evals)
plt.yscale('log')
plt.show()

# compress to 16 components like others
evecs = evecs[:,:16]

# compression is a linear operation
compressed_dv     =  np.array((dv_LFI)@evecs)

PCC = np.abs(np.corrcoef( compressed_dv,pars_LFI, rowvar = False))
makeplot(PCC, 'PCA-f')
```

# MOPED

Next is MOPED. We take the covariance matrix and derivatives with respect to the parameters at fiducial cosmology. We construct the MOPED compression matrix as outlined in the paper.

In short, MOPED finds linear compression that preserves the diagonal elements of the Fisher information matrix. This is achieved while enforcing orthonormality with respect to the fiducial covariance matrix.

As designed, we confirm that this compression diagonalizes the fiducial covariance matrix and conserves the determinant of the Fisher information matrix at fiducial cosmology.

However, the covariance matrix shows that the components of the compressed data vector are highly correlated. In the context of SBI this is a bad sign.

```python
# Compute covariance matrix from the simulated data vectors at fiducial
cov_matrix = np.cov(dv_cov.T)

# Compute parameter derivatives of the data vector with finite difference
deriv = dict()
for key in delta.keys():
    deriv[key] = np.mean((moped_dict[key][0]-moped_dict[key][1]),axis=0)/delta[key]
    
# the finite difference simulations vary sigma8 but we are more interested in S8. This corrects the derivative with chain rule
deriv['S8'] = deriv['delta_s8']/ 0.9309  + deriv['delta_Om']* 0.664

pars_ = ['S8' ,'delta_Om' ,'delta_w0','delta_A'  ,'delta_E'  ,'delta_Ob' ,'delta_ns' ,'delta_H0' ,'delta_dm1','delta_dm2','delta_dm3','delta_dm4','delta_dz1','delta_dz2','delta_dz3','delta_dz4' ]

deriv_array = np.array([deriv[key] for key in pars_])

# Compute MOPED vectors as defined in the paper
transf_matrix = np.zeros((len(pars_),(cov_matrix.shape[0])))
for count, p in enumerate(pars_):
    invcov_deriv = np.linalg.solve(cov_matrix, deriv[p])
    if count>0:
        coefs = transf_matrix[:count,:]@deriv[p]
        transf_matrix[count,:]= (invcov_deriv - transf_matrix[:count,:].T@coefs)/np.sqrt(invcov_deriv@deriv[p]- np.sum(coefs**2))
    else:
        transf_matrix[count,:]= (invcov_deriv)/np.sqrt(invcov_deriv@deriv[p])

# Compression is linear
compressed_dv        = np.matmul(transf_matrix, dv_LFI.T   ).T
compressed_dv_cov    = np.matmul(transf_matrix, dv_cov.T   ).T


# confirm that Fisher information is conserved
compressed_fisher = np.zeros((len(pars_),len(pars_)))
compressed_cov = np.cov(compressed_dv_cov.T)

for i, p in enumerate(pars_):
    for j, q in enumerate(pars_):
        compressed_fisher[i,j] = (transf_matrix@deriv[p]).T@np.linalg.solve(compressed_cov,(transf_matrix@deriv[q]))

full_fisher = np.zeros((len(pars_),len(pars_)))
for i, p in enumerate(pars_):
    for j, q in enumerate(pars_):
        full_fisher[i,j] = (deriv[p]).T@np.linalg.solve(cov_matrix,(deriv[q]))

print('Check that the Fisher information is preserved at fiducial cosmology')
print('compressed det F = ' + str(np.linalg.det(compressed_fisher)))
print('Full.      det F = ' + str(np.linalg.det(full_fisher)))

# Confirm that covmat is diagonalized 
plt.title("Covmat of the Compressed Fiducial Data Vectors")
plt.imshow(np.cov(compressed_dv_cov.T))
plt.colorbar(fraction=0.05, pad=0.03)
plt.show()

PCC = np.abs(np.corrcoef( compressed_dv,pars_LFI, rowvar = False))
makeplot(PCC, 'MOPED')
```

# e-MOPED

Here is e-MOPED, our proposed approach to MOPED. We compute the Jacobian and fiducial covariance matrix by assuming a linear model as outlined in the paper. With these, we construct the MOPED compression matrix.

Since this approach does not use simulations at fiducial cosmology or simulations at finite increments away from fiducial, the computational cost is vastly smaller. 

Unlike MOPED, e-MOPED does not diagonalize the fiducial covariance. But it does nearly diagonalize the covariance matrix of the compressed data vector. Furthermore, there is significant correlation between compressed data vector elements and the parameters of interest. This is a good sign for a dimensionality reduction technique for SBI.

```python
# The linear model assumptions hold better for sigma8 than for S8. 
pars_train = pars.copy()
pars_train[:,1] = pars_train[:,1]/np.sqrt(pars_train[:,0]/0.3)

# Compute the sampled parameter auto covariance, simulated data vector auto covariance
# and the parameter-data vector cross covariance
cov = np.cov(pars_train.T, dv.T)
cp = cov[:16,:16]
cd = cov[16:,16:]
cpd = cov[:16,16:]

# The solution to the linear regression implies a closed form solution for the Jacobian.
# This can be interpreted as an average Jacobian across the parameter space.
A = cpd.T@np.linalg.inv(cp)

# The linear model assumption implies that we should be able to 'slide' the data vectors to a fiducial point in parameter space. 
# With these 'fiducialized' data vectors we compute an estimate of the fiducial covariance matrix. 
p_me = np.mean(pars_train, axis = 0)
cov_matrix = np.cov((dv - (pars_train - p_me[np.newaxis,:])@A.T).T)

# We transform the Jacobian back to S8 coordinates, since ultimately we want to constrain S8. This is done with the chain rule.
A = A.T
A[1] = A[1]/ 0.9309  + A[0]* 0.664

pars_ = ['S8' ,'Om', 'w','A_IA'  ,'E_IA'  ,'Ob' ,'ns' ,'h','m1','m2','m3','m4','dz1','dz2','dz3','dz4' ]

deriv_array = A

# Compute MOPED vectors as defined in the paper
transf_matrix = np.zeros((len(pars_),(cov_matrix.shape[0])))
for count, p in enumerate(pars_):
    invcov_deriv = np.linalg.solve(cov_matrix, deriv_array[count])
    if count>0:
        coefs = transf_matrix[:count,:]@deriv_array[count]
        transf_matrix[count,:]= (invcov_deriv - transf_matrix[:count,:].T@coefs)/np.sqrt(invcov_deriv@deriv_array[count]- np.sum(coefs**2))
    else:
        transf_matrix[count,:]= (invcov_deriv)/np.sqrt(invcov_deriv@deriv_array[count])

# Compression is linear
compressed_dv        = np.matmul(transf_matrix, dv_LFI.T   ).T
compressed_dv_cov    = np.matmul(transf_matrix, dv_cov.T   ).T

# Ineed, e-MOPED does not diagonalize the fiducial covariance 
plt.title("Covmat of the Compressed Fiducial Data Vectors")
plt.imshow(np.cov(compressed_dv_cov.T))
plt.colorbar(fraction=0.05, pad=0.03)
plt.show()


PCC = np.abs(np.corrcoef( compressed_dv,pars_LFI, rowvar = False))
makeplot(PCC, 'e-MOPED')
```

# CCA

Last of the linear methods we present CCA, Canonical Correlation Analysis. CCA is equivalent to a generalized eigenvalue problem that maximizes mutual information between parameters and compressed data vector, which is how it is implemented here. 

The eigenvalues computed here are exp(2\*Mutual Information), and the plot of mutual information shows the first couple components carry most of the mutual information. 

The correlation matrices show that the compressed data vector components are uncorrelated to each other, but there is good correlation to the parameters of interest. 

```python
# sigma8, not S8, has a more linear relationship to the data vectors. 
pars_train = pars.copy()   
pars_train[:,1] = pars_train[:,1]/np.sqrt(pars_train[:,0]/0.3)


# Compute the sampled parameter auto covariance, simulated data vector auto covariance
# and the parameter-data vector cross covariance
cov = np.cov(pars_train.T, dv.T)
cp = cov[:16,:16]
cd = cov[16:,16:]
cpd = cov[:16,16:]

# This 'cl' can be understood as the projection of 'cp' to data vector space
cl = cpd.T@np.linalg.inv(cp)@cpd

# As seen in the paper, this generalized eigenvalue problem is equivalent to CCA
# but is more numerical stable as 'cd' and 'cd-cl' are both invertible.
# This problem is motivated as mutual information maximization under Gaussian linear model assumptions
evals, evecs = scipy.linalg.eigh(cd, cd - cl)

# In the context of the CCA, only min( dim(param), dim(data vector) ) components are real and the rest are noise. 
evals = evals[::-1][:16]
evecs = evecs[:,::-1][:,:16]

plt.title("Mutual Information Plot for CCA")
plt.xlabel("Compressed Data Vector (CDV) Index")
plt.ylabel("Mutual Information Between Parameter and CDV")
plt.plot(np.log(evals)/2)
plt.show()

compressed_dv = dv_LFI@evecs

PCC = np.abs(np.corrcoef( compressed_dv,pars_LFI, rowvar = False))
makeplot(PCC, 'CCA')
```

# NN-MSE with Optuna hyperparameter tuning

We consider NN-MSE as a representative non-linear method. NN-MSE compresses the data vectors by attempting to infer the original parameters from the data vectors. This is done by training a neural network using a mean squared error (MSE) loss function. This compresses the data vector to a vector only as long as the parameter vector. 

Each parameter has a different optimal architecture since the complexity of each parameter's impact on the data vector is different. So, in this case, we would need to train 16 different neural networks of varying architectures, and concatenate their outputs to get a 16 dimensional compressed data vector. In this example, we train a neural network to infer the 0th parameter, S8.

Here is an implementation with Optuna to tune the hyperparameters describing the network architecture. Optuna adjusts the architecture and trains the corresponding neural network until the validation loss of the optimized neural network model is minimized with respect to architecture. 

```python
# NN-MSE with Optuna requires more packages

# import the hyperparameter tuning package, Optuna
import optuna

# In this example, we run purely on CPUs
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# Import necessary tensorflow assets
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.layers import Dense, Dropout, LeakyReLU, ReLU
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.layers import Input, Add
from tensorflow.keras.callbacks import EarlyStopping

# In this example, the neural network is being trained to infer S8, the 0th parameter         
p = 0

len_dv = dv.shape[1]  
samples = dv.shape[0]

# split the simulated data vectors to training & validation to train the neural network
rng = default_rng()
numbers = rng.choice(samples , size=3000, replace=False)
special = np.in1d(np.arange(samples ),numbers)


pars_train = np.array(pars[~special])
pars_val =  np.array(pars[special])
dv_train = np.array(dv[~special])
dv_val = np.array(dv[special])

# We trim the data vectors so that 98% of the data is between -0.5 and 1.5, 1% below -0.5, and 1% above 1.5.
# This is because neural networks are most effective with values between 0 and 1. 
# The outliers below -0.5 and above 1.5 are also clipped at -3 and 4, so there are no crazy outliers.
lo  = np.percentile(dv_train, q = 1 ,axis=0)
hi  = np.percentile(dv_train, q = 99,axis=0)

min_ = -3; max_ = 4

# Training, validation, and DV to be transformed once training is done. 
optuna_input = dict()
optuna_input['DV_train']  = np.clip((dv_train-lo)*2/(hi-lo)-0.5,min_,max_)
optuna_input['DV_val']    = np.clip((dv_val  -lo)*2/(hi-lo)-0.5,min_,max_)
optuna_input['DV_test']   = np.clip((dv_LFI  -lo)*2/(hi-lo)-0.5,min_,max_)
optuna_input['DV_cov']    = np.clip((dv_cov  -lo)*2/(hi-lo)-0.5,min_,max_)
optuna_input['params_train'] = pars_train
optuna_input['params_val']   = pars_val
optuna_input['params_test']  = pars_LFI


# Function to create a neural network given the layer sizes and learning rate
def create_model(input_size, layer_sizes, lr):
    model = Sequential()

    model.add(Dense(layer_sizes[0], input_dim=input_size, kernel_initializer='normal'))
    model.add(LeakyReLU(alpha=0.01))

    for size in layer_sizes[1:]:
        model.add(Dense(size, kernel_initializer='normal'))
        model.add(LeakyReLU(alpha=0.01))

    model.add(Dense(1, kernel_initializer='normal')) 
    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    return model


# Optuna's loss function is the optimal validation loss for a given architecture
def objective(trial):
    input_size = optuna_input['DV_train'].shape[1]  # Example to get input_size
    num_layers = trial.suggest_int('num_layers', 1, 6)  # Let Optuna decide between 1 and 10 layers


    # Start with a large possible range for the first layer
    layer_sizes = [trial.suggest_int('layer_0_size', 20, 120)]
    # Subsequent layers have a max size smaller than the previous layer
    for i in range(1, num_layers):
        max_size = layer_sizes[i-1] - 1  # Ensure strictly smaller size
        layer_sizes.append(trial.suggest_int(f'layer_{i}_size', 5, max(max_size, 20)))  # Minimum size 5   

    lr = trial.suggest_float('learning_rate', 1e-5, 1e-3, log=True)

    model = create_model(input_size, layer_sizes, lr)

    # load the data vectors
    X_train, X_val =  np.array(optuna_input['DV_train']),          np.array(optuna_input['DV_val'])
    y_train, y_val =  np.array(optuna_input['params_train'][:,p]), np.array(optuna_input['params_val'][:,p])

    # Fit the given neural network
    history = model.fit(
        X_train, y_train, epochs=100, batch_size=32, verbose=0,
        validation_data=(X_val, y_val),
        callbacks=[EarlyStopping(monitor='val_loss', patience=8, verbose=0, mode='min', restore_best_weights=True)]
    )
    predictions = model.predict(X_val)
    
    # Return the best loss value.
    print ('training loop loss ',np.mean((y_val-predictions)**2))
    return copy.deepcopy(np.mean((y_val-predictions)**2))
    
    
def doit(p):
    
    # Optuna operates by managing a sqlite data base with information about past experiments and architectures
    storage = optuna.storages.RDBStorage(url='sqlite:////' + current_dir + '_{0}.db'.format(p), engine_kwargs={"connect_args": {"timeout": 200}})

    study = optuna.create_study(storage=storage, load_if_exists=True, direction='minimize',study_name ='{0}'.format(p))


    # Run the architecture optimization
    total_trials = 100
    batch_size = 50
    for _ in range(0, total_trials, batch_size):
        study.optimize(objective, n_trials=batch_size, n_jobs=1)


    print('Best parameters:', study.best_params)
    print('Best validation loss:', study.best_value)
    

    
    # Rebuild and rerun the model with the best parameters
    input_size = optuna_input['DV_train'].shape[1]
    best_params = study.best_params

    best_model = create_model(
        input_size,  # You need to specify this based on your actual data setup
        [best_params[f'layer_{i}_size'] for i in range(best_params['num_layers'])],
        best_params['learning_rate'])


    X_test,X_train, X_val =  np.array(optuna_input['DV_test']),           np.array(optuna_input['DV_train']),           np.array(optuna_input['DV_val'])
    y_test,y_train, y_val =  np.array(optuna_input['params_test'][:,p]),  np.array(optuna_input['params_train'][:,p]),  np.array(optuna_input['params_val'][:,p])


    # use the same used for the training just to check everything is OK 
    history_final = best_model.fit(
            X_train, y_train, epochs=150, batch_size=32, 
            validation_data=(X_val, y_val),
            callbacks=[EarlyStopping(monitor='val_loss', patience=8, mode='min', restore_best_weights=True)]
        )


    predictions = best_model.predict(X_val)

    MSE = np.mean((y_val-predictions)**2)

    results = dict()
    results['predictions_val'] = predictions
    results['p_val'] = y_val
    results['MSE_10k_val'] = np.mean((y_val-predictions)**2)


    
    # Transform the cov and DV_LFI data vectors for use with SBI
    predictions = best_model.predict(optuna_input['DV_cov'])

    results['cov'] = predictions


    predictions = best_model.predict(X_test)

    results['predictions_test'] = predictions
    results['p_test'] = y_test
    results['MSE_test'] = np.mean((y_test-predictions)**2)
    results['best_params'] = study.best_params
    
    # Save result of compression
    np.save(current_dir+'_optuna_{0}.npy'.format(p),results)


doit(p)
```

This is a plot of the initial input values of $S_8$ against the predicted values of $S_8$. The Mean squared error is 0.0098 and $R^2$ is 0.72. 

Neural networks are also quite good at predicting $\Omega_m, \eta_{\rm IA}, A_{\rm IA}$ as well. The other parameters such as $n_s, \Omega_b$ are a bit harder for neural networks as their impact on the data vector is fainter or more complicated. 

We repeat the above for each parameter, and concatenate the predicted parameter values to use as the compressed data vectors for SBI. 

```python
y = np.load(current_dir+'_optuna_{0}.npy'.format(0), allow_pickle = True).item()

from sklearn.metrics import r2_score

r2 = r2_score(y['p_test'], y['predictions_test'])

plt.scatter(y['p_test'], y['predictions_test'], label = 'MSE = ' + str(round(y['MSE_test'],4)) + '\n'+ r'$R^2$    = ' + str(round(r2,2)), s = 0.5)
plt.xlabel("Real Value")
plt.ylabel("Value Predicted by NN")
plt.title("Result of Optuna-enabled NN-MSE in Predicting S8")
plt.plot((0.12,0.55),(0.12,0.55), color = 'red')
plt.legend()
plt.show()
```
